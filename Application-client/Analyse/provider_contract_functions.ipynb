```python
import matplotlib.pyplot as plt
import csv
```

```python
# ETH
# ---

x_axis_ProviderContractInitialisation = []
y_axis_ProviderContractInitialisation = []

x_axis_GettingSelfInfo = []
y_axis_GettingSelfInfo = []

x_axis_GetOwner = []
y_axis_GetOwner = []

x_axis_AddProvider = []
y_axis_AddProvider = []

# POLYGON
# -------

x_axis_ProviderContractInitialisation_pol = []
y_axis_ProviderContractInitialisation_pol = []

x_axis_GettingSelfInfo_pol = []
y_axis_GettingSelfInfo_pol = []

x_axis_GetOwner_pol = []
y_axis_GetOwner_pol = []

x_axis_AddProvider_pol = []
y_axis_AddProvider_pol = []
```

```python
with open("./providers_ethereum.csv", 'r') as f:
    data = csv.reader(f,delimiter=',')

    for dataline in data:
        if(dataline[0] == "ProviderContractInitialisation&Connection"):
            x_axis_ProviderContractInitialisation.append(dataline[3])
            y_axis_ProviderContractInitialisation.append(float(dataline[1]))
        elif(dataline[0] == "GettingSelfInfo"):
            x_axis_GettingSelfInfo.append(dataline[3])
            y_axis_GettingSelfInfo.append(float(dataline[1]))
        elif(dataline[0] == "GetOwner-Contract"):
            x_axis_GetOwner.append(dataline[3])
            y_axis_GetOwner.append(float(dataline[1]))
        elif(dataline[0] == "AddProvider"):
            x_axis_AddProvider.append(dataline[3])
            y_axis_AddProvider.append(float(dataline[1]))


with open("./providers_polygon.csv", 'r') as f:
    data = csv.reader(f, delimiter=',')

    for dataline in data:
        if(dataline[0] == "ProviderContractInitialisation&Connection"):
            x_axis_ProviderContractInitialisation_pol.append(dataline[3])
            y_axis_ProviderContractInitialisation_pol.append(float(dataline[1]))
        elif(dataline[0] == "GettingSelfInfo"):
            x_axis_GettingSelfInfo_pol.append(dataline[3])
            y_axis_GettingSelfInfo_pol.append(float(dataline[1]))
        elif(dataline[0] == "GetOwner-Contract"):
            x_axis_GetOwner_pol.append(dataline[3])
            y_axis_GetOwner_pol.append(float(dataline[1]))
        elif(dataline[0] == "AddProvider"):
            x_axis_AddProvider_pol.append(dataline[3])
            y_axis_AddProvider_pol.append(float(dataline[1]))
```

```python
# ProviderContractInitialisation
plt.plot(x_axis_ProviderContractInitialisation, y_axis_ProviderContractInitialisation,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_ProviderContractInitialisation_pol, y_axis_ProviderContractInitialisation_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Provider Contract Initialisation (ETH/POLYGON)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# GettingSelfInfo
plt.plot(x_axis_GettingSelfInfo, y_axis_GettingSelfInfo,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_GettingSelfInfo_pol, y_axis_GettingSelfInfo_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Getting Self Info (ETH/POLYGON)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# AddProvider
plt.plot(x_axis_AddProvider, y_axis_AddProvider,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_AddProvider_pol, y_axis_AddProvider_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Add Provider (ETH/POL)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```
